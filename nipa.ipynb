```python
import keras
import numpy as np
from keras.applications import vgg16
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
vgg_model = vgg16.VGG16(weights = 'imagenet')
```

```python
filename = 'nipa_image_sample.jpg'  # 직접 이미지 파일 하나를 준비해준다
```

```python
org = load_img(filename,  # 이미지 로딩하기
               target_size = (224,224))  # 해당 사이즈로 로딩해준다. vgg16이 224*224 사이즈를 쓰기 때문에.

img = img_to_array(org) # 이미지를 넘파이 어레이로 바꿔준다

plt.imshow(np.uint8(img))  #unit8 안 쓰면 스머프로 나온다
```

```python
def load_nipa_img(num):

    nipa_image = load_img(f'/volumes/KTWusb/nipa/test/{num}.jpg', target_size = (224,224))  # 해당 사이즈로 로딩해준다. vgg16이 224*224 사이즈를 쓰기 때문에.
    img = img_to_array(nipa_image) # 이미지를 넘파이 어레이로 바꿔준다

#    plt.imshow(np.uint8(img))  #unit8 안 쓰면 스머프로 나온다
    
    return img
```
